```python
import networkx as nx
import numpy as np
import pandas as pd
import random as rm
import seaborn as sns
import matplotlib.pyplot as plt
#from colorspacious import cspace_converter
#import matplotlib as mpl
#import matplotlib.pyplot as plt
```

# PRIMER MODELO (Grafo simple)

```python
#generamos una matriz de n√∫mero aleatorios
G = nx.DiGraph()
nmax = 9
nodes = ['Bc', 'B1', 'B2','B3','B4', 'S1','S2']
colores = ['gold','limegreen','limegreen','limegreen','limegreen','lightgray','lightgray']
nnodes = len(nodes)
M = [[rm.randint(1,11)%nmax for j in range(nnodes)] for i in range(nnodes)]
```

```python
M
```

```python
palette = sns.color_palette("hot", nmax)
palette
```

```python
#G.add_edge(1,2,color='r',weight=2)
#nx.draw(G, pos, edges=edges, edge_color=colors, width=weights)
```

```python
for i in range(nnodes):
    for j in range(nnodes):
        if M[i][j] != 0 and i != j and j != 0 and i != 5 and i != 6:
            G.add_edge(nodes[i],nodes[j],color=palette[M[i][j]],weight=M[i][j]*0.5)

edges = G.edges()
colors = [G[u][v]['color'] for u,v in edges] #edges=edges
weights = [G[u][v]['weight'] for u,v in edges]
nx.draw_circular(G, with_labels=True, node_color=colores, node_size=500, font_size=10, font_color='black', font_weight='bold', edge_color=colors)#, width=weights)
```
